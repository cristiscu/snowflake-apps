#### Generate fake data

```python
from faker import Faker
f = Faker()
output = [[f.name(), f.address(), f.city(), f.state(), f.email()]
    for _ in range(1000)]
output
```

#### Connect to Snowflake

```python
import os
from snowflake.snowpark import Session

pars = {
    "account": os.environ['SNOWFLAKE_ACCOUNT'],
    "user": os.environ['SNOWFLAKE_USER'],
    "password": os.environ['SNOWFLAKE_PASSWORD'],
    "database": "test",
    "schema": "public"
}
session = Session.builder.configs(pars).create()
session.get_current_warehouse()
```

#### Create Snowpark DataFrame

```python
schema = ["name", "address", "city", "state", "email"]
df = session.create_dataframe(output, schema=schema)
df.show()
```

#### Save in table and check back data

```python
df.write.mode("overwrite").save_as_table("customers_fake2")
session.sql('select * from customers_fake2').show()
```
